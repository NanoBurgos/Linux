## **Para prender la luz de fondo del teclado externo en ubuntu 22.04, seguí estos pasos: -**

**PASO 1** Método para cambiar de Wayland a X este paso es obligatorio

Editá el archivo de configuración,

sudo nano /etc/gdm3/custom.conf

para descomentar [sacar el # de] la línea que dice

WaylandEnable=false

guardá el archivo (ctrl + x y después y ('yes') u otra tecla si no usás inglés y finalmente Enter) y reiniciá.

**PASO 2** Método Uno

abrí una terminal

sudo xmodmap -e 'add mod3 = Scroll_Lock'

poné tu contraseña

apretá la tecla de bloqueo de desplazamiento (Scroll Lock) para ver si se prende la luz de fondo

abrí una terminal

xset led on
